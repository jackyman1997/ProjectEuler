```python
tri = []  # for store
for i in text: 
    i = list(map(int, i[:-1].split()))
    tri.append(i)
tri
```

```python
# greed 
path = []
pos = []
for i in tri: 
    
    if len(i) == 1: 
        maxi = max(i)
        posit = i.index(maxi)
        
        path.append(maxi)
        pos.append(posit)
    
    else: 
        maxi = max(i[pos[-1]],i[pos[-1]+1])
        posit = i.index(maxi)
        
        path.append(maxi)
        pos.append(posit)
    
print(path,end='')
sum(path)
```

```python
# greed backward
path2 = []
pos = []
for i in reversed(tri): 
    
    if len(i) == len(tri): 
        maxi = max(i)
        posit = i.index(maxi)
        
        path2.append(maxi)
        pos.append(posit)
        
    elif len(i) == 1: 
        path2.append(i[0])
        pos.append(0)
        
    elif pos[-1] != len(i)-1 and pos[-1] != 0: 
        maxi = max(i[pos[-1]],i[pos[-1]-1])
        posit = i.index(maxi)
        
        path2.append(maxi)
        
        if posit == pos[-1] or posit == pos[-1]+1: 
            pos.append(posit)
        
        else:
            if maxi == i[pos[-1]-1]:
                pos.append(pos[-1])
            if maxi == i[pos[-1]]:
                pos.append(pos[-1])
                
    else: 
        pos.append(pos[-1]-1)
        maxi = i[pos[-1]-1]
        path2.append(maxi)
        
print(path2[::-1],end='')
sum(path2)
```
