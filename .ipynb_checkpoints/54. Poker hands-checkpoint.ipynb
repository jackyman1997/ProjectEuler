## 54. Poker hands
In the card game poker, a hand consists of five cards and are ranked, from lowest to highest, in the following way:

- High Card: Highest value card.  
- One Pair: Two cards of the same value.  
- Two Pairs: Two different pairs.  
- Three of a Kind: Three cards of the same value.  
- Straight: All cards are consecutive values.  
- Flush: All cards of the same suit.  
- Full House: Three of a kind and a pair.  
- Four of a Kind: Four cards of the same value.  
- Straight Flush: All cards are consecutive values of same suit.  
- Royal Flush: Ten, Jack, Queen, King, Ace, in same suit.  

The cards are valued in the order:  
2, 3, 4, 5, 6, 7, 8, 9, 10, Jack, Queen, King, Ace.  

If two players have the same ranked hands then the rank made up of the highest value wins; for example, a pair of eights beats a pair of fives (see example 1 below). But if two ranks tie, for example, both players have a pair of queens, then highest cards in each hand are compared (see example 4 below); if the highest cards tie then the next highest cards are compared, and so on.

Consider the following five hands dealt to two players:
    
| Hand  | Player 1 | Player 2 | Winner | 
|:-----:|:--------:|:--------:|:------:|  
| 1     | 5H 5C 6S 7S KD<br>Pair of Fives | 2C 3S 8S 8D TD<br>Pair of Eights | Player 2 |  
| 2     | 5D 8C 9S JS AC<br>Highest card Ace | 2C 5C 7D 8S QH<br>Highest card Queen | Player 1 |
| 3     | 2D 9C AS AH AC<br>Three Aces | 3D 6D 7D TD QD<br>Flush with Diamonds | Player 2 |
| 4     | 4D 6S 9H QH QC<br>Pair of Queens<br>Highest card Nine | 3D 6D 7H QD QS<br>Pair of Queens<br>Highest card Seven|Player 1 |
| 5     | 2H 2D 4C 4D 4S<br>Full House With Three Fours |3C 3D 3S 9S 9D<br>Full House with Three Threes | Player 1 |


The file, <a href='https://projecteuler.net/project/resources/p054_poker.txt'>poker.txt</a>, contains one-thousand random hands dealt to two players. Each line of the file contains ten cards (separated by a single space): the first five are Player 1's cards and the last five are Player 2's cards. You can assume that all hands are valid (no invalid characters or repeated cards), each player's hand is in no specific order, and in each hand there is a clear winner.

I need a class to identify what 'rank' is player 1 or 2 having in each round

```python
class poker():
    
    def __init__(self, string: str):
        self.cards = string.split(' ')
        self.values = ''
        self.suits = ''
        for card in self.cards:
            self.values += card[0]
            self.suits += card[1]
        # need to numberise the 'T' 'K' 'Q' 'J' 'A'
        self.ValueOrder = '23456789TJQKA'
        self.SuitOrder = 'SHCD'
        # need a sort in values
        numberise = []
        for value in self.values:
            numberise.append(self.ValueOrder.index(value))
        numberise.sort()
        # back in sorted string format 
        self.values = ''
        for value in numberise:
            self.values += self.ValueOrder[value]
        pass
    
    def checker(self):
        # remember ranked
        self.rank = {
            'High Card': 0, 
            'One Pair': 0, 
            'Two Pairs': 0, 
            'Three of a Kind': 0, 
            'Straight': 0, 
            'Flush': 0, 
            'Full House': 0, 
            'Four of a Kind': 0, 
            'Straight Flush': 0, 
            'Royal Flush': 0 
        }
        # checking repeated cards
        self.repeat = {}
        for card in self.values:
            self.repeat[card] = self.repeat.get(card, 0) + 1
        # ranks can be determined by the len of the self.repeat dictionary 
        if len(self.repeat) == 4:
            self.rank['One Pair'] = 1
        if len(self.repeat) == 3:
            # check Two Pairs or Three of a Kind
            if self.repeat[max(self.repeat, key=self.repeat.get)] == 2:
                self.rank['Two Pairs'] = 1
            elif self.repeat[max(self.repeat, key=self.repeat.get)] == 3:
                self.rank['Three of a Kind'] = 1
        if len(self.repeat) == 2: 
            # check full house or Four of a Kind 
            if self.repeat[max(self.repeat, key=self.repeat.get)] == 3:
                self.rank['Full House'] = 1
            elif self.repeat[max(self.repeat, key=self.repeat.get)] == 4:
                self.rank['Four of a Kind'] = 1
        # now if len is 5, the hard part
        if len(self.repeat) == 5:
            # set Flush and Straight bool
            Flush, Straight = False, False
            if self.values in self.ValueOrder: # if substring of ValueOrder
                Straight = True
            if len( set(self.suits) ) == 1: # if only one repeated element in suits
                Flush = True
            # identify the remaining
            if Flush and Straight: 
                if self.values == 'TJQKA':
                    self.rank['Royal Flush'] = 1
                else:
                    self.rank['Straight Flush'] = 1
            if Flush:
                self.rank['Flush'] = 1
            if Straight:
                self.rank['Straight'] = 1
            if not (Flush or Straight):
                self.rank['High Card'] = 1
        return max(self.rank, key=self.rank.get)        
```

I also need another function to determine which player wins

```python
def win(hand):
    ranking = [
        'High Card', 
        'One Pair', 
        'Two Pairs', 
        'Three of a Kind', 
        'Straight', 
        'Flush', 
        'Full House', 
        'Four of a Kind', 
        'Straight Flush', 
        'Royal Flush'
    ]
    
    p1, p2 = hand[0], hand[1]
    
    if ranking.index(p1.checker()) > ranking.index(p2.checker()):
        winner = '1'
    else:
        winner = '2'
        
    if ranking.index(p1.checker()) == ranking.index(p2.checker()):
        '''
        I cheated in here, 
        i know there are no same ranked hands other than:
        'High Card' and 'One Pair'
        so in here i only do the comparison of these two kinds
        '''
        p1_values = p1.values
        p2_values = p2.values
        ValueOrder = '23456789TJQKA'
        
        # High Card
        if p1.checker() == 'High Card':
            while p1_values[-1] == p2_values[-1]:
                p1_values.pop(-1)
                p2_values.pop(-1)

            if ValueOrder.index(p1_values[-1]) > ValueOrder.index(p2_values[-1]):
                winner = '1'
            else:
                winner = '2'
        
        # One Pair
        if p1.checker() == 'One Pair':
            # check pair
            for i in range(len(p1_values)-1):
                if p1_values[i] == p1_values[i+1]:
                    p1_pair = p1_values[i]
                if p2_values[i] == p2_values[i+1]:
                    p2_pair = p2_values[i]
            
            if ValueOrder.index(p1_pair) > ValueOrder.index(p2_pair):
                winner = '1'
            else:
                winner = '2'
            
            # if same pair, check High Card
            if p1_pair == p2_pair:
                # remove the pair from the cards 
                p1_HighCard = p1_values.replace(p1_pair, '')
                p2_HighCard = p2_values.replace(p2_pair, '')
            
                while p1_HighCard[-1] == p2_HighCard[-1]:
                    p1_HighCard.pop(-1)
                    p2_HighCard.pop(-1)

                if ValueOrder.index(p1_HighCard[-1]) > ValueOrder.index(p2_HighCard[-1]):
                    winner = '1'
                else:
                    winner = '2' 
    
    return winner 
```

```python
# libs
from _myeulertools import doc_import

# input
q54_input = {'filename': 'p054_poker.txt'}

# function
def q54(filename: str):
    # import .txt
    text = doc_import('p054_poker.txt')
    
    # splitting player 1 and 2
    players = []
    for hand in text.split('\n')[:-1]: # for some reason the last element is ['', ''] so ignore
        players.append( [ hand[:14], hand[15:] ] )
    
    # using a list to store player info (poker obj)
    hands = []
    for player in players:
        hands.append(
            [poker(player[0]), poker(player[1])]
        )
        
    # count how many wins player 1 has
    cnt = 0
    for hand in hands:
        if win(hand) == '1':
            cnt += 1
    
    return print('player 1 won %i hands'%(cnt))
```
